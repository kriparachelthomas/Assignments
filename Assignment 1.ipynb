# 5. Perform meaningful visualizations using the dataset

# 1) KDE Plot

```python
dataset[['PW','PL']].plot.kde()
plt.title('plot between Petal width and petal length')
```

```python
dataset[['SW','SL']].plot.kde()
plt.title('plot between Sepal width and Sepal length')
```

A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

# 2)Scatter plot

```python
plt.figure(figsize=(6,4))
plt.scatter(dataset['SW'],dataset['SL'],s=10)
plt.title('Plot of SW VS SL',fontsize=16)
plt.style.use('classic')
plt.grid()
plt.xlabel('SW')
plt.ylabel('SL')
```

```python
plt.figure(figsize=(6,4))
plt.scatter(dataset['PW'],dataset['PL'],s=10)
plt.title('Plot of PW VS PL',fontsize=16)
plt.style.use('classic')
plt.grid()
plt.xlabel('PW')
plt.ylabel('PL')
```

# 3) Histogram

```python
plt.figure(figsize=(6,6))
plt.hist(dataset['SL'],color='red',rwidth=0.9)
plt.title('Distribution of SL',fontsize=16)
plt.xlabel('SL')
plt.ylabel('Frequency')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(6,6))
plt.hist(dataset['SW'],color='blue',rwidth=0.9)
plt.title('Distribution of SW',fontsize=16)
plt.xlabel('SW')
plt.ylabel('Frequency')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(6,6))
plt.hist(dataset['PL'],color='green',rwidth=0.9)
plt.title('Distribution of PL',fontsize=16)
plt.xlabel('PL')
plt.ylabel('Frequency')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(6,6))
plt.hist(dataset['PW'],color='yellow',rwidth=0.9)
plt.title('Distribution of PW',fontsize=16)
plt.xlabel('PW')
plt.ylabel('Frequency')
plt.xticks()
plt.yticks()
```
