# 3. Plot a bar graph of male vs female buyers participated in the sales.


```python
car['Buyer Gender'].value_counts()
```

```python
plt.figure(figsize=(7,6))
labels = ['Male','Female']
sns.set_style("darkgrid")
sns.countplot(x='Buyer Gender', data=car,hue='Buyer Gender')
plt.title("Count of Male and Female Buyers)",fontsize=12)
plt.xlabel("BUYER GENDER",fontsize=12)
plt.ylabel("COUNT",fontsize=12)
plt.xticks()
plt.yticks()
```

 Findings: There are more female buyers than male buyers.
